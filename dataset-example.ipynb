Example of "real-life" data and jupyter notebooks

```python
#public libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("data/auto-mpg.csv")
print(data.head())
print("Are there any empty cells?: ", data.isnull().values.any())
```

```python
data.plot(figsize=(20,10))
```

```python
data_X = data.iloc[:,1:6]
print(data_X.head())
```

```python
data_y = data.iloc[:,0]
print(data_y.head())
```

```python
data_X = data_X.to_numpy()
data_y = data_y.to_numpy()
print(data_X.shape, data_y.shape)
```

SKLEARN does not like (397,) format, reshape it to (397,1)

```python
data_y = data_y[:,np.newaxis]
print(data_y.shape)
```

```python
#split into train test
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=0.8)#, random_state=42)

#make the model
reg = LinearRegression().fit(X_train, y_train)
prediction = reg.predict(X_test)
for x, y, real in zip(X_test, prediction, y_test):
    print(x, y, real)
print("R^2 ON TRAINING SET: ", reg.score(X_test, y_test))
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
fig, ax = plt.subplots()
fig.tight_layout(pad=2.5, w_pad=1.5, h_pad=0) #fix margins
ax.scatter(X_test[:,0], y_test, c="blue")
ax.scatter(X_test[:,0], prediction, c="red")
ax.set(xlabel='X', ylabel='Y')
ax.set_title('Cylinders vs MPG')
plt.show()
```
